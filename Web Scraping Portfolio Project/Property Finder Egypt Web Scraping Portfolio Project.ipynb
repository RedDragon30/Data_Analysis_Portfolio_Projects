```python
# Import Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
type = []
price = []
description = []
location = []
bedroom = []
bathroom = []
area = []
released = []
page_num = 1
```

```python
while True:

    try:
        url = f'https://www.propertyfinder.eg/ar/search?c=2&fu=0&rp=m&ob=mr&page={page_num}'# Properties for rent in Egypt
        url2 = f'https://www.propertyfinder.eg/ar/search?c=1&fu=0&ob=mr&page={page_num}'# Properties for sale in Egypt (optional)
        
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36", 
                    "X-Amzn-Trace-Id": "Root=1-689332fc-71130940703cf209648cd329"}
        
        page = requests.get(url, headers=headers)    
        soup = BeautifulSoup(page.content, 'lxml')
        
        page_limit_element = soup.find('span', {'aria-label':"Search results count"})
    
        if page_limit_element:
            number_text = page_limit_element.text.split(' ')[0].replace(',', '')  
            if number_text.isdigit():
                page_limit = int(number_text)         
            else:
                print("No valid results found, stop the process.")
                break
        else:
            print("No results found for search, stop process.")
            break  
        if (page_num > page_limit // 25):
            print('Pages ended, terminate')
            break

        # Types of Properties
        types = soup.find_all('p', {'class':"styles-module_content__property-type__QuVl4"})
        types = types if types else None

        # Property Prices
        prices = soup.find_all('p', {'class':"styles-module_content__price__SgQ5p"})
        prices = prices if prices else None

        # Property Descriptions
        descriptions = soup.find_all('h2', {'class':"styles-module_content__title__eOEkd"})
        descriptions = descriptions if descriptions else None

        # Property Locations
        locations = soup.find_all('p', {'class':"styles-module_content__location__bNgNM"})
        locations = locations if locations else None

        # Property Bedrooms
        bedrooms = soup.find_all('p', {'class':"styles-module_content__details-item__mlu9B", 'data-testid':"property-card-spec-bedroom"})
        bedrooms = bedrooms if bedrooms else None

        # Property Bathrooms
        bathrooms = soup.find_all('p', {'class':"styles-module_content__details-item__mlu9B", 'data-testid':"property-card-spec-bathroom"})
        bathrooms = bathrooms if bathrooms else None

        # Areas
        areas = soup.find_all('p', {'class':"styles-module_content__details-item__mlu9B", 'data-testid':"property-card-spec-area"})
        areas = areas if areas else None

        # Property Released
        released_ = soup.find_all('p', {'class':"styles-module_footer__publish-info__UVabq"})
        released_ = released_ if released_ else None

        for i in range(len(bedrooms)):
            type.append(types[i].text if types[i] else None)
            price.append(prices[i].text if prices[i] else None)
            description.append(descriptions[i].text if descriptions[i] else None)
            location.append(locations[i].text if locations[i] else None)
            bedroom.append(bedrooms[i].text if bedrooms[i] else None)
            bathroom.append(bathrooms[i].text if bathrooms[i] else None)
            area.append(areas[i].text if areas[i] else None)
            released.append(released_[i].text if released_[i] else None)
        
        print(f'page {page_num} scraped')   
        page_num += 1
        time.sleep(1)

    except:
        print('Error occurred.')
```

```python
data = {
    'Type':type,
    'Area':area,
    'Bedroom':bedroom,
    'Bathroom':bathroom,
    'Released':released,
    'Location':location,
    'Description':description,
    'Price':price
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('Properties.csv', index=False, encoding='utf-8-sig')
```
